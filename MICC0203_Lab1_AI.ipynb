```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Training data
X = np.array([[1], [2], [3], [4]])
y = np.array([2, 4, 6, 8])

# Create a linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X, y)

# Predict using the trained model
X_test = np.array([[5]])
y_pred = model.predict(X_test)

print(y_pred)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Training data
X = [[2.0, 1.0], [1.0, 3.0], [3.0, 2.0], [3.0, 4.0]]
y = [0, 0, 1, 1]

# Create a logistic regression model
model = LogisticRegression()

# Fit the model to the training data
model.fit(X, y)

# Predict using the trained model
X_test = [[1.0, 2.0], [4.0, 3.0]]
y_pred = model.predict(X_test)

print(y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Training data
X = [[2.0, 1.0], [1.0, 3.0], [3.0, 2.0], [3.0, 4.0]]
y = [0, 0, 1, 1]

# Create a KNN classifier with k=3
model = KNeighborsClassifier(n_neighbors=3)

# Fit the model to the training data
model.fit(X, y)

# Predict using the trained model
X_test = [[1.0, 2.0], [4.0, 3.0]]
y_pred = model.predict(X_test)

print(y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB

# Training data
X = [[2.0, 1.0], [1.0, 3.0], [3.0, 2.0], [3.0, 4.0]]
y = [0, 0, 1, 1]

# Create a Naive Bayes classifier
model = GaussianNB()

# Fit the model to the training data
model.fit(X, y)

# Predict using the trained model
X_test = [[1.0, 2.0], [4.0, 3.0]]
y_pred = model.predict(X_test)

print(y_pred)
```

```python
from sklearn.decomposition import PCA

# Training data
X = [[2.0, 1.0], [1.0, 3.0], [3.0, 2.0], [3.0, 4.0]]

# Create a PCA transformer
pca = PCA(n_components=1)

# Fit the transformer to the training data and transform it
X_transformed = pca.fit_transform(X)

print(X_transformed)
```

```python
from sklearn.cluster import KMeans

# Training data
X = [[2.0, 1.0], [1.0, 3.0], [3.0, 2.0], [3.0, 4.0]]

# Create a K-Means clustering model with 2 clusters
model = KMeans(n_clusters=2)

# Fit the model to the training data
model.fit(X)

# Predict cluster labels for the training data
labels = model.labels_

print(labels)
```

```python
from sklearn.neural_network import MLPClassifier

# Training data
X = [[0, 0], [0, 1], [1, 0], [1, 1]]
y = [0, 1, 1, 0]

# Create a multi-layer perceptron classifier with 2 hidden layers
model = MLPClassifier(hidden_layer_sizes=(3, 2))

# Fit the model to the training data
model.fit(X, y)

# Predict using the trained model
X_test = [[0, 1], [1, 0]]
y_pred = model.predict(X_test)

print(y_pred)
```
